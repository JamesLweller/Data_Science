Firstly, what sex is most likely to carry out mass murder?

```r
Gender <- factor(mass$Gender)
```

```r
Gender <- fct_collapse(Gender,
                  "Male"=c("M","Male"),
                  "Male/Female"=c("M/F","Male/Female"),
                  "Unknown"="Unknown",
                  "Female"="Female")
```

```r
plot(Gender,
     main = "Sex of the Perpetrator",
    ylab = "Number of the Perpetrator",
    xlab = "Sex",
    col = "dark red",
    ylim = c(0,300))
```
