# Incident Of Mass shooting based on Dates

Is there a increase in mass shooting over the past 50 years. Looking at first the year, then the month and finally the day of the month.

```r
year <- str_sub(mass$Date,1, 4)
year <- as.numeric(year)
years <- c(1960:2020)
hist(year, 
     col = "grey", 
     main = "Year",
     ylim = c(0, 70),
     breaks = years)
```

```r
table(year)
```

This histogram shows that mass shooting in america have increased since 1966. With 2015 and 2016 having the most shooting in the past 50 years.

```r
#Month
month <- str_sub(mass$Date, 6, 7)
month <- as.numeric(month)
months <- c(0:12)
hist(month, 
     col = "grey", 
     main = "Month",
     ylim = c(0, 60),
     breaks=months)
```

```r
table(month)
```

This histogram looks at each month from this it would appear that February followed by March has had the most mass shooting the past 51 years. ALthough, it would appear that there are less shooting in the summer months.

```r
#Day
day <- str_sub(mass$Date,9,10)
day <- as.numeric(day)
days <- c(0:31)
hist(day, 
     col = "grey", 
     main = "Day", 
     ylim = c(0, 25),
     breaks=days)
```

This histogram shows that there is no particular day that mas shooting are likely to occur.

# Location off Mass Shooting

```r
#longitude and latitude to new data frame
lon <- mass$Longitude
lat <- mass$Latitude
co <- as.data.frame(cbind(lon,lat))
```

```r
#Remove N/A from the data frame
co <- na.omit(co)
```

```r
#Download googlemap 
shooting_map <- get_map(location = c(lon = mean(co$lon), lat = mean(co$lat)), zoom = 4,
                      maptype = "satellite", scale = 2)

#Plot the longitude and latitude on the map
ggmap(shooting_map) +
  geom_point(data = co, aes(x = lon, y = lat, fill = "red", alpha = 0.8), size = 4, shape = 21) +
  guides(fill=FALSE, alpha=FALSE, size=FALSE, na.rm = TRUE)
```
