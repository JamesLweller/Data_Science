# Big Five Personality Traits

## Introduction
The purpose of this project is to elevate the big five personality traits (Openness to experience, Conscientiousness, Extraversion, Agreeableness, Neuroticism) https://www.verywell.com/the-big-five-personality-dimensions-2795422. In relation to gender difference between males and females, race and hand uses.

The second part of this project will look at the relationship between race, gender and personailty traits.

The third part of this project will look at the relationship between participants dominant hand, gender and personailty trait.

This data comes from the Open Source Psychometrics Project (https://openpsychometrics.org).

This dataset consists of 19720 participants (11985 females and 7608 males).

Raw file http://openpsychometrics.org/_rawdata/BIG5.zip

Website https://openpsychometrics.org/tests/IPIP-BFFM/

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(101)
df = pd.read_csv("/home/james/Downloads/BIG5/data.csv", header=0, sep="\s+")
df = df.dropna()
```

```python
df.head()
```

The scores for each of the five traits need to be summed into a new total score.

There are four different variables used for gender (1=Male, 2=Female, 3=Other 0=missed). This project is only looking at males and females, the other two variables (Other and missed) were removed using the subset function.

```python
df['Extraversion'] = df.iloc[:, 7:17].sum(axis=1)
df['Neuroticism'] = df.iloc[:, 17:27].sum(axis=1)
df['Agreeableness'] = df.iloc[:, 27:37].sum(axis=1)
df['Conscientiousness'] = df.iloc[:,37:47].sum(axis=1)
df['Openness'] = df.iloc[:, 47:57].sum(axis=1)
```

```python
male = df[df.gender == 1]
female = df[df.gender == 2]
```

```python
female['gender'] = 'F'
male['gender'] = 'M'
```

```python
df = pd.concat([male, female], ignore_index=True)
```

## Gender difference in personality

These five density plots show personality differences between males and females. The blue overlay is male and the red overlay is female.

The lines on each graph represent the mean score (Males are blue and Females are red). Note: higher or lower score on each item do not signify positive or negative traits.

### Openness to experience

Openness to experience is a trait associated with insight, imagination, creativity and adventure.

https://www.psychologistworld.com/influence-personality/openness-to-experience-trait

```python
sns.kdeplot(male['Openness'], shade=True, color='b', alpha=.3, legend=False)
sns.kdeplot(female['Openness'], shade=True, color='r', alpha=.2, legend=False)
plt.axvline(male['Openness'].mean(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(female['Openness'].mean(), color='r', linestyle='dashed', linewidth=1)
plt.title('Openness')
plt.xlabel('Openness')
plt.ylabel('Density')
plt.legend(('Male', 'Female'),
           loc='upper right')
plt.show()
```

### Conscientiousness

https://www.psychologistworld.com/influence-personality/conscientiousness-personality-trait

## Race & Personality Traits

The next part of this project looks at the relationship between race, gender and personality trait. The dataset recorded 9 different races were collected Arctic (Siberian, Eskimo), Caucasian(European, Indian, Middle East, North African), Indigenous Australian, Native American, North East Asian (Mongol, Tibetan, Korean Japanese), Pacific (Polynesian, Micronesian), South East Asian (Chinese, Thai, Malay, Filipino), West African, Bushmen, Ethiopian. Two further races where recored Mixed race and Other, finally, some participants choose to not included their race recorded as missing.

```python
df['race'] = df['race'].replace({0:'missed',
1:'Mixed Race', 
2:'Arctic (Siberian, Eskimo)', 
3:'Caucasian (European)', 
4:'Caucasian (Indian)', 
5:'Caucasian (Middle East)', 
6:'Caucasian (North African, Other)', 
7:'Indigenous Australian', 
8:'Native American', 
9:'North East Asian (Mongol, Tibetan, Korean Japanese, etc)', 
10:'Pacific (Polynesian, Micronesian', 
11:'South East Asian (Chinese, Thai, Malay, Filipino, etc)', 
12:'West African, Bushmen, Ethiopian', 
13:'Other'})
```

### Extraversion

```python
sns.boxplot(x='race', y='Extraversion', data=df, palette='Set1')
plt.title('Extraversion')
plt.xticks(rotation=90)
plt.show()
```

This Barplot indicates that there is little extraversion scores for each of the races. However, based on the scores Caucasian (European) have the lowest levels of extraversion with a mean score of 30.38. Whereas, Arctic (Siberian, Eskimo) scored the highest for extraversion with a mean score of 33.15.

### Neuroticism

```python
sns.boxplot(x='race', y='Neuroticism', data=df, palette='Set1')
plt.title('Neuroticism')
plt.xticks(rotation=90)
plt.show()
```

This boxplot indicates that Arctic (Siberian, Eskimo) have the lowest levels of Neuroticism with a mean of 27.46, this would suggest that Arctic (Siberian, Eskimo) have lower levels of anxiety and depression (https://www.psychologytoday.com/us/basics/neuroticism). However, other data suggests that other white Caucasian, Eskimo where the second most lilkey to use mental health services (https://www.samhsa.gov/data/sites/default/files/MHServicesUseAmongAdults/MHServicesUseAmongAdults.pdf)(P1). Whereas, Caucasian (Indian) have the highest levels of Neuroticism with a mean of 32.58.

### Agreeableness

```python
sns.boxplot(x='race', y='Agreeableness', data=df, palette='Set1')
plt.title('Agreeableness')
plt.xticks(rotation=90)
plt.show()
```

This data indicates that Caucasian (European) have the lowest levels of Agreeableness (31.53). Whereas, Arctic (Siberian, Eskimo) have the highest levels of Agreeableness (34.30)

### Conscientiousness

```python
sns.boxplot(x='race', y='Conscientiousness', data=df, palette='Set1')
plt.title('Conscientiousness')
plt.xticks(rotation=90)
plt.show()
```

The boxplot would suggest that there is very little difference between races interms of Conscientiousness. Although, Indigenous Australian (29.68) scored the highest, Whereas, South East Asian (Chinese, Thai, Malay, Filipino) (31.88) scored the highest on the Conscientiousness trait.

### Openness to experience

```python
sns.boxplot(x='race', y='Openness', data=df, palette='Set1')
plt.title('Openness')
plt.xticks(rotation=90)
plt.show()
```

The final Trait Openness to experience indicates that there is a significant amount of overlap between races. However, Caucasian (Middle East) scored the lowest (32.25). Whereas, Caucasian (European) scored the highest (33.65) for Openness to experience.

## Personailty Traits & hand preference

This examines the differences between males and females for each of the personality traits in relation to hand preference.

```python
df['hand'] = df['hand'].replace({0:'missed',
1:"Right", 
2:"Left", 
3:"Both"})
```

### Extraversion

```python
sns.boxplot(x='hand', y='Extraversion', hue='gender', data=df, palette='Set1')
plt.title('Extraversion')
plt.xticks(rotation=90)
plt.show()
```

The first boxplot indicates that males scored higher then females for Extraversion. Furthermore, both males and females that had no prefered hand had higher levels of extraversion compared to individuals that had a prefered hand.

### Neuroticism

```python
sns.boxplot(x='hand', y='Neuroticism', hue='gender', data=df, palette='Set1')
plt.title('Neuroticism')
plt.xticks(rotation=90)
plt.show()
```

This boxplot indicates that females scored higher then males for Neuroticism. Although, there was no difference between males and females for individuals that had no prefered hand. Furthermore, females that had no prefer hand had the lowest levels of Neuroticism.

### Agreeableness

```python
sns.boxplot(x='hand', y='Agreeableness', hue='gender', data=df, palette='Set1')
plt.title('Agreeableness')
plt.xticks(rotation=90)
plt.show()
```

This boxplot indicates that left handed males are the most Agreeableness (29.92) where as left handed females have the lowest levels of Agreeableness (32.09). Furthermore, females had higher levels of Agreeableness compared to males.

### Conscientiousness

```python
sns.boxplot(x='hand', y='Conscientiousness', hue='gender', data=df, palette='Set1')
plt.title('Conscientiousness')
plt.xticks(rotation=90)
plt.show()
```

This boxplot indicates that both males and females with no hand preference scored higher compared to the other groups. Whereas, right handed individuals scored the lowest for Conscientiousness.

### Openness to experience

```python
sns.boxplot(x='hand', y='Openness', hue='gender', data=df, palette='Set1')
plt.title('Openness')
plt.xticks(rotation=90)
plt.show()
```

This boxplot indicates that both males and females that had no hand preference scored higher for Openness to experience compared to the other groups, Furthermore there was much overlap between the other three groups.
